### imports

```python
import numpy as np
from math import comb
import matplotlib.pyplot as plt
import math
```

### Binomial
n number of trials with a binary outcome. E.g: 10 coin tosses --> n=10, p=0.5

```python
def binomial(n, p):
    q = 1-p

    probs = []
    for x in range(n+1):
        pmf = comb(n,x) * (p**x) * ((q)**(n-x))
        probs.append(pmf)

    plt.figure(figsize=(4,2))
    plt.bar(np.arange(0,n+1),probs)
    plt.show()
    print(f'mean (n*p): {n*p}')
    print(f'variance (n*p*q): {n*p*q}')
```

```python
binomial(n=20, p=0.5)
```

```python
binomial(n=20, p=0.3)
```

### Poisson
probability of a given number of events occuring in a fixed period of time

e.g: on average, 5 people enter a store in a given day

```python
def poisson(x):
    probs = []
    for i in range(0,2*x):
        pmf = x**i * np.e**-x / math.factorial(i)
        probs.append(pmf)
        
    plt.figure(figsize=(4,2))
    plt.bar(np.arange(0,2*x),probs)
    plt.show()
    print(f'mean (x): {x}')
    print(f'variance (x): {x}')
        
```

```python
poisson(5)
```

```python
poisson(30)
```

### Pareto
The distribution behind the 80/20 rule. the top 20% are responsible for 80%.

e.g: wealth of people in a society. 20% of the people own more than 80% of the wealth

```python
def pareto(a,xm):
#     alpha = a -> shape of the curve
    probs = []
    for i in range(0,20):
        if i < xm:
            pdf = 0
        else:
            pdf = a*xm**a / i**(a+1)
        probs.append(pdf)
        
    plt.figure(figsize=(4,2))
    plt.bar(np.arange(0,20),probs)
    plt.show()
        
```

```python
pareto(0.2,1)
```

```python
pareto(1,1)
```

```python
def uniform(a,b):
    probs = []
    for i in range(a,b):
        if i >= a and i <= b:
            pdf = 1/ (b-a)
        else:
            pdf = 0
        probs.append(pdf)
        
    plt.figure(figsize=(4,2))
    plt.bar(np.arange(0,20),probs)
    plt.show()
        
```

```python
uniform(0,20)
```

```python
def normal(m,s):
    probs = []
    for i in range(-100,100):
        
        pdf = (np.e**(-0.5*((i-m)/s)**2)) / (s*(2*np.pi)**0.5)
        probs.append(pdf)
        
    plt.figure(figsize=(4,2))
    plt.bar(np.arange(-100,100),probs)
    plt.show()
```

```python
normal(0,10)
```

```python
normal(0,20)
```

```python
normal(20,30)
```

### multinomial
similar to binomial but generalized to greater than 2 possible outcomes

```python
# number of trials
n = 20
# probability of each outcome
p = [0.1,0.4,0.5]
# most likely outcome (based on n & p)
k = [2,8,10]
```

in the code below, we will keep k[2] as 10, and we will play around with k[0] and k[1]

```python
pmfs = []

for i in range(n-k[2]+1):
    k[0] = i
    k[1] = n - k[2] - i

    combs = math.factorial(n) / (math.factorial(k[0])*math.factorial(k[1])*math.factorial(k[2]))
    pmf =  combs * p[0]**k[0] * p[1]**k[1] * p[2]**k[2]
    pmfs.append(pmf)
    
plt.bar(np.arange(0,11),pmfs)
plt.title('likelihood of different possible values of k[0] given it\'s p=0.1')
plt.xlabel('k[0]')
plt.show()
```

the above likelihood looks similar to a binomail distribution of n=20 & p=0.1
